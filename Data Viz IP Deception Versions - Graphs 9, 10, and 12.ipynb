### Graph 9:

**A. Original vs. Deception version:**

Original graph's claim: Homicide numbers have decreased in the past two decades.

Rebuttal presented by deception version: In the past two decades, homicide numbers have gone up and down and is now rising sharply.  

There are several key differences that lead to the differences between the two versions.  First, raw numbers were used instead of rates per 100,000 people, so now the y-axis spread is different. It's also important to note that now we no longer have context for the overall population, which can lead to deception and jumping to false conclusions.  We don't know if the population is growing or shrinking, we only know the number of firearm homicides.  Secondly, I used the timeframe 1995 to 2015 instead of their timeframe.  This more accurately represents the "last two decades" that the original claim refers to, and this is the key change that led to big differences in the graph.  This makes reveals that there were ups and downs in the past two decades and that now the number of firearm homicides is increasing again.  Thirdly, the y-axis is curtailed, as it begins far above 0.  As a result, the y-axis scale is also stretched out, which makes the ups and downs seem a lot bigger than they actually are.

### Graph 12:

**A. Original vs. Deception versions:**

Original claim: Guns allow people to commit suicide more easily.

Rebuttal presented by deception version: Poison actually allows people to try and commit suicide more easily, as there are far more suicide attempts using poison than using a gun.  This may be because substances such as drugs are more accessible than guns.

The original visual used data from Indiana 1990 to 1997, but this deception version uses data from a published article by R.S. Spicer et al. (2000), which can be found here: https://ajph.aphapublications.org/doi/pdf/10.2105/AJPH.90.12.1885
The study includes suicide fatalities and attempts pooled from 8 different states’ hospital records from 1989 to 1997.  Since the data is from 8 states rather than one state, the results shown by the graph are slightly different.  In addition, the original pie charts displayed the data labels in percentages, but the deception pie charts display them as the raw counts so we can see how many attempts and fatalities there actually were for each suicide method. This version shows that the total poison attempts were 65305, whereas the firearm attempts were only 929.  This suggests that it is easier to try and commit suicide using poison rather than firearms, contrary to the original claim.

In addition to changing the labels, the pie charts are exploded and have shadows added to skew their appearance.  More importantly, the pie charts alternate color schemes.  For instance, the firearm pie chart has dark orange for attempted and yellow for fatal, but the suffocation pie chart has yellow for attempted and dark orange for fatal.  If the reader does not look carefully at the labels, they might not realize this and be comparing apples to oranges.  

**B. Making of/Documentation:**

The data was obtained from the aforementioned study, which directly provided the raw counts for each suicide method.  These numbers were entered into an Excel CSV and read into Python.  The matplotlib library was used to generate all four pie charts.  An extra function was defined to change the automatic percentage labels to raw counts.

**C. Future enhancements:**

The rebuttal might be more convincing and obvious if these pie charts were all consolidated into one bar graph instead, with four bars, each representing the attempted suicide count for each suicide method.  That way, the bars will be side by side and easier to compare.

```python
suicides = pd.read_csv("suicide_methods.csv")
suicides
```

```python
#create labels

labels = ['fatal', 'attempted']

poison = [1462, 65305]

firearm = [4365, 929]

cut = [158, 12949]

suff = [1405, 888]
```

```python
#create custom function to return raw counts instead of percentages in pie charts
explode=(0, 0.2)
def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return '{v:d}'.format(p=pct,v=val)
    return my_autopct


plt.pie(firearm, labels=labels, shadow=True, autopct=make_autopct(firearm), explode = explode, startangle = 90, colors = ['#F9E79F', '#D35400'])
plt.title("Firearm Suicides")
plt.axes().set_aspect(aspect = 'equal')

plt.show()
```

```python
explode=(0, 0.1)
plt.pie(suff, labels=labels, shadow=True, autopct=make_autopct(suff), explode = explode, startangle = 30, colors = ['#D35400', '#F9E79F'])
plt.title("Suffocation Suicides")
plt.axes().set_aspect(aspect = 'equal')

plt.show()
```

```python

plt.pie(poison, labels=labels, autopct=make_autopct(poison), shadow=True, explode = explode, startangle = 90, colors = ['#D35400', '#F9E79F'])
plt.title("Poison Suicides")
plt.axes().set_aspect(aspect = 'equal')

plt.show()
```

```python

explode=(0, 0.3)
plt.pie(cut, labels=labels, autopct=make_autopct(cut), shadow=True, explode = explode, startangle = 90, colors = ['#F9E79F', '#D35400'])
plt.title("Cut Suicides")
plt.axes().set_aspect(aspect = 'equal')

plt.show()
```
