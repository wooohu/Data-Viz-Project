```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
guns = pd.read_csv("guns_2013.csv")
```

```python
guns
```

```python
#create scatterplot

plot1 = sns.regplot(x=guns["death_rate"], y=guns["gun_rate"], color="red")

for i, txt in enumerate(guns.state):
    plot1.annotate(txt, (guns.death_rate[i],guns.gun_rate[i]))

plt.title("Gun Ownership vs. Gun Deaths")
plt.xlabel("Gun ownership (percent of adults)")
plt.ylabel("Gun deaths per 100,000 residents")

sns.plt.show()
```
