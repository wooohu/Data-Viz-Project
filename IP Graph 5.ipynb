```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
guns = pd.read_csv("guns_2013.csv")
```

```python
guns

#rename column:
guns = guns.rename(columns={'death_rate.1': 'death_rate'})

guns
```

In the dataframe, death_rate represents num_deaths/total_pop.  This dataset includes people age > 18 from 2013 only.

```python
#create lists of states for partisan composition
dem = ['WA', 'OR', 'NV', 'CA', 'CO', 'NM', 'MN', 'IL', 'WV', 'NY', 'VT', 'ME', 'MD', 'DE', 'NJ', 'CT', 'RI', 'MA']
other = ['NE', 'IA', 'KY', 'VA', 'NH']
#else = republican

#create variable for each state's political party
def party(row):
    if row['state'] in dem:
        return 'D'
    elif row['state'] in other:
        return 'O'
    else:
        return 'R'

guns['party'] = guns.apply(lambda row: party(row),axis=1)

guns.head(5)
```

```python
#create scatterplot
sns.set_style("white")

def color(row):
    if row['party'] == 'D':
        return '#0000FF'
    elif row['party'] == 'R':
        return '#FF0000'
    else:
        return '#90EE90'

guns['color'] = guns.apply(lambda row: color(row),axis=1)


plot1 = sns.regplot(x="gun_rate", y="death_rate", data = guns, scatter_kws = {'s':100, 'facecolors': guns['color']})

#sns.regplot(data=df, x="x", y="y", scatter_kws={'facecolors':df['color']})

for i, txt in enumerate(guns.state):
    plot1.annotate(txt, (guns.gun_rate[i],guns.death_rate[i]), size='large', color='black', weight='semibold')

plt.title("Gun Ownership vs. Gun Deaths")
plt.xlabel("Gun ownership (percent of adults)")
plt.ylabel("Gun deaths (percent of adults)")


#x1,x2,y1,y2 = plt.axis()

#plt.axis((0, 100, 0, 0.03))

# Move the legend to an empty part of the plot
plt.legend(loc='lower right')


sns.plt.show()
```
